# KDE

```python
sns.kdeplot(data=data,
            x="New cases",
           hue="WHO Region",)
plt.show()
```

```python
sns.kdeplot(data=data,
            x="Recovered",
           hue="WHO Region",fill= True,
           common_norm=False,
           palette="crest",alpha=1,linewidth=0)
plt.show()
```
